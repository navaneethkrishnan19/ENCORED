```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from CSV file
df = pd.read_csv("Train_Data.csv")

# Grouping by gender and taking the mean of numerical columns
grouped_data = df.groupby('gender').mean()

# Plotting
grouped_data.plot(kind='line', marker='o')
plt.title('Average Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.legend(title='Score Categories')
plt.grid(True)
plt.show()
```
